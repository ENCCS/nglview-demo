```python
import nglview as nv

pyridine_xyz = """11

C       -0.180226841      0.360945118     -1.120304970
C       -0.180226841      1.559292118     -0.407860970
C       -0.180226841      1.503191118      0.986935030
N       -0.180226841      0.360945118      1.29018350
C       -0.180226841     -0.781300882      0.986935030
C       -0.180226841     -0.837401882     -0.407860970
H       -0.180226841      0.360945118     -2.206546970
H       -0.180226841      2.517950118     -0.917077970
H       -0.180226841      2.421289118      1.572099030
H       -0.180226841     -1.699398882      1.572099030
H       -0.180226841     -1.796059882     -0.917077970
"""
```

```python
import qcelemental as qcel

mol = qcel.models.Molecule.from_data(pyridine_xyz)

v0 = nv.show_qcelemental(mol)

c1 = v0.add_component("HOMO.cube")
c1.clear()

c1.add_surface(isolevelType="value", isolevel=0.02, color="red")
c1.add_surface(isolevelType="value", isolevel=-0.02, color="blue")

v0
```

```python
from openbabel import openbabel

obConversion = openbabel.OBConversion()
obConversion.SetInAndOutFormats("xyz", "pdb")

mol = openbabel.OBMol()
obConversion.ReadString(mol, pyridine_xyz)


with open("pyridine.pdb", "w") as f:
    f.write(obConversion.WriteString(mol))

v1 = nv.show_file("pyridine.pdb")

c1 = v1.add_component("HOMO.cube")
c1.clear()

c1.add_surface(isolevelType="value", isolevel=0.02, color="red")
c1.add_surface(isolevelType="value", isolevel=-0.02, color="blue")

v1
```
